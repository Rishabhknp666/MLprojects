```python
from scipy.stats import boxcox
temp_data= []
temp_data= pd.DataFrame(temp_data)

for i in ndf.columns:
    plt.figure(figsize=(10, 5))
    c = np.array(ndf[i])

    # Check for constant values
    if np.all(c == c[0]):
        print(f"Skipping {i} - constant values present.")
        c=pd.DataFrame(c)
        temp_data = pd.concat([temp_data,c],axis=1)
        continue

    # Check for non-positive values
    if np.any(c <= 0):
        print(f"Skipping {i} - contains zero or negative values.")
        c=pd.DataFrame(c)
        temp_data = pd.concat([temp_data,c],axis=1)
        continue

    working_data, lambda_value = boxcox(c + 1)  # Adding 1 to handle zero values
    
    working_data = pd.DataFrame(working_data)
    temp_data = pd.concat([temp_data,working_data],axis=1)
    
    sns.kdeplot(working_data, color="blue")
    plt.title(f'KDE Plot after Box-Cox Transformation for {i}')
    plt.show()
```
