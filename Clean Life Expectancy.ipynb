```python
from sklearn.preprocessing import PowerTransformer

temp_data = pd.DataFrame()  # Initialize an empty DataFrame

for i in ndf.columns:
    plt.figure(figsize=(10, 5))
    c = np.array(ndf[i])

    # Check for constant values
    if np.all(c == c[0]):
        print(f"Skipping {i} - constant values present.")
        c = pd.DataFrame(c)
        temp_data = pd.concat([temp_data, c], axis=1)
        continue

    # Check for non-positive values
    if np.any(c <= 0):
        print(f"Skipping {i} - contains zero or negative values.")
        c = pd.DataFrame(c)
        temp_data = pd.concat([temp_data, c], axis=1)
        continue

    # Initialize PowerTransformer with Yeo-Johnson method
    yeo_johnson_transformer = PowerTransformer(method='yeo-johnson')

    # Fit and transform the data
    working_data = yeo_johnson_transformer.fit_transform(c.reshape(-1, 1))

    # Convert transformed data to DataFrame
    working_data = pd.DataFrame(working_data)

    # Plot the KDE plot after Yeo-Johnson Transformation
    sns.kdeplot(working_data[0], color="blue")
    plt.title(f'KDE Plot after Yeo-Johnson Transformation for {i}')
    plt.show()

    # Concatenate transformed data to temp_data DataFrame
    temp_data = pd.concat([temp_data, working_data], axis=1)

# Assign column names
temp_data.columns = ndf.columns
```
