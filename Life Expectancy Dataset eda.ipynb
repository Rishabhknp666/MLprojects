```python
from scipy.stats import boxcox

for i in tdf.columns:
    plt.figure(figsize=(10, 5))
    c = np.array(tdf[i])

    # Check for constant values
    if np.all(c == c[0]):
        print(f"Skipping {i} - constant values present.")
        continue

    # Check for non-positive values
    if np.any(c <= 0):
        print(f"Skipping {i} - contains zero or negative values.")
        continue

    tdf2, lambda_value = boxcox(c + 1)  # Adding 1 to handle zero values

    sns.kdeplot(tdf2, color="blue")
    plt.title(f'KDE Plot after Box-Cox Transformation for {i}')
    plt.show()
```

```python
tdf1 = pd.DataFrame()
plt.figure(figsize=(10, 7))
for i in ndfip.columns:
        
    # Clip outliers to a wider range
    clipped_values = ndfip[i].clip(ndfip[i].quantile(0.05), ndfip[i].quantile(0.95))
    
    tdf1[i] = clipped_values
    # Create a boxplot with the clipped values
    sns.boxplot(x=clipped_values)
    
    plt.title(f'Boxplot after Outlier Transformation for {i}')
    plt.show()
```

```python
tdf1 = pd.DataFrame()
for i in ndfip.columns:
    plt.figure(figsize=(10, 7))
    
    column_data = ndfip[i]

    # Calculate the 5th and 95th percentiles to define the outlier threshold
    lower_bound = column_data.quantile(0.05)
    upper_bound = column_data.quantile(0.95)
    
    outliers = (column_data < lower_bound) | (column_data > upper_bound)
    
    column_data[column_data < lower_bound] = lower_bound
    column_data[column_data > upper_bound] = upper_bound
    
    sns.boxplot(x=column_data)
    
    plt.title(f'Boxplot after Outlier Transformation for {i}')
    plt.show()
```
